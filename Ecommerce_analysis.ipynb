```python
# Corrected method for counting occurrences of values in the 'state' column
top_states = df['State'].value_counts().nlargest(5).index

# Display the counts
print(top_states)
```

```python
#shipping mode
```

```python
df['Ship Mode'].value_counts()
```

```python
 x = df['Ship Mode'].value_counts().index
 y = df['Ship Mode'].value_counts().values
print(x)
print(y)
```

```python
plt.figure(figsize=(12, 8))
plt.pie(y,labels = x, autopct = "%0.2f%%")
plt.title("shipping mode ")
plt.legend()
plt.show()
```

```python
# bivariate analysuis 
```

```python
plt.figure(figsize=(12, 8))
palette = {'Furniture': 'magenta', 'Office Supplies': 'orange', 'Technology': 'blue'}
sns.countplot(x='Ship Mode', data=df, hue='Category', palette=palette)

plt.show()
```

```python
top_states_df = df[df['State'].isin(top_states)]
plt.figure(figsize=(12, 8))
sns.countplot(x = 'State', data = df )
plt.title('State', fontsize = 25)
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```
